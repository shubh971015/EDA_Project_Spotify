```python

df.drop_duplicates(keep='first',inplace=True).
```

```python
# Calculate the percentage of duplicate rows in the dataset
duplicate_percentage = round(((9999 - df.drop_duplicates().shape[0]) / 9999) * 100, 2)

# Explanation:
# - calculates the percentage of duplicate rows by comparing the original number of rows (9999) 
#   with the number of unique rows after dropping duplicates, 
#   then rounds the result to 2 decimal places.

# Print the percentage of duplicate data in the dataset
print(f"This dataset has {duplicate_percentage}% duplicate data")
```

```python
# To calculate the mean popularity of songs with the same track name and artist name

group_data=df.groupby(['Track Name','Artist Name(s)'])['Popularity'].mean().reset_index()
group_data
```

```python

# Sort the dataset by the 'Popularity' column in descending order and select the top 5 entries
top_5_popular = group_data.sort_values(by='Popularity', ascending=False).head(5)

# Set the figure size before creating the plot
plt.figure(figsize=(15, 15))

# Create a bar plot with 'Artist Name(s)' on the x-axis and 'Popularity' on the y-axis
ax = top_5_popular.plot(kind='bar', x='Artist Name(s)', y='Popularity', legend=False)

# Display the plot title and labels
plt.title('Top 5 Artists by Popularity')
plt.ylabel('Popularity Score')
plt.xlabel('Artist Name(s)')
plt.xticks(rotation=45, ha='right')  # Rotate the artist names for better visibility

# Adding bar values on top of the bars
for index, value in enumerate(top_5_popular['Popularity']):
    plt.text(index, value + 1, str(value), ha='center', va='bottom')

#plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```
